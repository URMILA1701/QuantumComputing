# Problem 1: Quantum Teleportation (Bell State)

### Part A- Qiskit Implementation

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
```

##### Step 1: Arbitrary Qubit Preparation (3 Marks)

```python
qc=QuantumCircuit(3,1)
qc.ry(1.3, 0)
qc.rz(2.1, 0)
```

##### Step 2: Create a Bell Pair

```python
qc.h(1)
qc.cx(1,2)
```

##### Step 3: Bell Measurement Circuit

```python
qc.cx(0,1)
qc.h(0)
```

##### Step4: Fully Quantum Correction

```python
qc.cx(1,2)
qc.cz(0,2)
```

##### Step 5: Undo the Unitary U

```python
qc.rz(-2.1, 2)
qc.ry(-1.3, 2)
```

##### Step 6: Measurement and Histogram

###### 6.1 Measuring Bob’s qubit

```python
qc.measure(2,0)
```

###### 6.2 Run the circuit with at least 5000 shots

```python
backend = Aer.get_backend('aer_simulator')
job = backend.run(qc, shots=5000)
result = job.result()

counts = result.get_counts()
counts
```

###### 6.3 Plot the histogram

```python
plot_histogram(counts)
```

###### 6.4 full circuit diagram.

```python
qc.draw('mpl')
```
